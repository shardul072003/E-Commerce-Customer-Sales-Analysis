```python
import pandas as pd

print("Original shape:", df.shape)

df = df.dropna(subset=['CustomerID', 'Description'])
print("After dropping nulls:", df.shape)

df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
print("After removing cancellations:", df.shape)

df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
print("After removing bad quantities/prices:", df.shape)

df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df['CustomerID'] = df['CustomerID'].astype(int)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

print("\nFinal cleaned shape:", df.shape)
print(df.dtypes)
```

```python
from datetime import datetime
```

```python
# referecne date one day after the last transaction
reference_date = datetime(2012, 12, 10)
```

```python
# calculate rfm (recencey , frequnecy , monetary)
rfm = df.groupby('CustomerID').agg(
    Recency = ('InvoiceDate' , lambda x: (reference_date - x.max()).days),
    Frequency = ('InvoiceNo' , 'nunique'),
    Monetary =('TotalPrice' , 'sum')).reset_index()

rfm['Monetary'] = rfm['Monetary'].round(2)
```

```python
print(rfm.shape)
```

```python
print(rfm.head(10))
```

```python
# higher recency score = bought more recently
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=4 , labels=[4,3,2,1])
```

```python
# higher Frequency Score = bought more often
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first') , q=4 , labels=[1,2,3,4])
```

```python
# higher Monetary Score = spent more
rfm['M_score'] = pd.qcut(rfm['Monetary'].rank(method='first'), q=4 , labels=[1,2,3,4])
```

```python
# Combine Score
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
```

```python
print(rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary', 'R_Score', 'F_score', 'M_score', 'RFM_Score']].head(10))
```

```python
def assign_segment(score):
    if score in ['444', '443', '434', '344', '433']:
        return 'Champion'
    elif score in ['442', '441', '432', '431', '342', '341']:
        return 'Loyal'
    elif score in ['334', '333', '332', '331', '324', '323']:
        return 'Potential Loyalist'
    elif score in ['412', '411', '311', '312', '411']:
        return 'New Customer'
    elif score in ['244', '243', '234', '233', '144', '143']:
        return 'At Risk'
    elif score in ['111', '112', '121', '122', '211', '212']:
        return 'Lost'
    else:
        return 'Others'

rfm['Segment'] = rfm['RFM_Score'].apply(assign_segment)
```

```python
# drop all duplicate and category columns and start fresh
rfm = rfm[['CustomerID', 'Recency', 'Frequency', 'Monetary']].copy()

# make sure Monetary is float
rfm['Monetary'] = rfm['Monetary'].astype(float)

# recreate scores cleanly
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=4, labels=[4, 3, 2, 1]).astype(int)
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4]).astype(int)
rfm['M_Score'] = pd.qcut(rfm['Monetary'].rank(method='first'), q=4, labels=[1, 2, 3, 4]).astype(int)

# combine into RFM score
rfm['RFM_Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)

# apply segments
rfm['Segment'] = rfm['RFM_Score'].apply(assign_segment)

# verify dtypes
print(rfm.dtypes)
print("\n")

# count per segment
print(rfm['Segment'].value_counts())
print("\n")

# summary per segment
print(rfm.groupby('Segment').agg(
    Total_Customers=('CustomerID', 'count'),
    Avg_Recency=('Recency', 'mean'),
    Avg_Frequency=('Frequency', 'mean'),
    Avg_Monetary=('Monetary', 'mean')
).round(2))
```
